```python
df['fixed acidity'] = df['fixed acidity'].fillna(df['fixed acidity'].mean())
df['volatile acidity'] = df['volatile acidity'].fillna(df['volatile acidity'].mean())
df['citric acid'] = df['citric acid'].fillna(df['citric acid'].mean())
df['residual sugar'] = df['residual sugar'].fillna(df['residual sugar'].mean())
df['chlorides'] = df['chlorides'].fillna(df['chlorides'].mean())
df['pH'] = df['pH'].fillna(df['pH'].mean())
df['sulphates'] = df['sulphates'].fillna(df['sulphates'].mean())
```

```python
df.isna().sum()
```

```python
df['quality'].unique()
```

```python
df.head(50)
```

```python
df['goodquality'] = [-1 if x < 6 else 0 if x == 6 else 1 for x in df['quality']]
```

```python
df['type'].unique()
```

```python
df['type'] = df['type'].map({"red":0,"white":1})
```

```python
X = df.drop(['goodquality','id','type'],axis=1)
y = df['goodquality']
```

```python
X.head()
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
X_scaled = pd.DataFrame(X_scaled, columns=X.columns)
X_scaled
```

```python
plt.figure(figsize=(20,25), facecolor='white')
plotnumber = 1

for column in X_scaled:
    if plotnumber<=15 :     # as there are 9 columns in the data
        ax = plt.subplot(5,3,plotnumber)
        sns.distplot(X_scaled[column])
        plt.xlabel(column,fontsize=20)
        #plt.ylabel('Salary',fontsize=20)
    plotnumber+=1
plt.show()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_scaled,y, test_size=0.2, random_state=42)
```

```python
rfc = RandomForestClassifier(n_estimators=200)
rfc.fit(X_train,y_train)
```

```python
y_pred = rfc.predict(X_test)
y_pred
```

```python
sv = SVC()
sv.fit(X_train,y_train)
```

```python
sv_pred = sv.predict(X_test)
print(sv_pred)
```

```python
from sklearn.metrics import classification_report, confusion_matrix, f1_score
clf = classification_report(y_test,y_pred)
print(clf)
```

```python
f1 = f1_score(y_test,y_pred,average='weighted')
print(f1)
```

```python
cm = confusion_matrix(y_test,y_pred)
print(cm)
```
